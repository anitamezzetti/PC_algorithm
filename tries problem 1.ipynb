```python
di_G.clear
G.clear

G = nx.Graph()
G.add_edge(1,2)
G.add_edge(1,2)
G.add_edge(3,2)
G.add_edge(3,4)
G.add_edge(1,4)
G.add_edge(3,1)
G.add_edge(5,4)

poss = nx.planar_layout(G)
nx.draw(G, pos=poss, with_labels=True)
```

```python
di_G = nx.DiGraph()
di_G.add_nodes_from(G.nodes)
di_G.add_edge(2,1)
di_G.add_edge(3,2)
di_G.add_edge(4,3)

nx.draw(di_G, pos=poss, with_labels=True)
```
